```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('/content/sample_data/UCI_Credit_Card.csv')
df
print(df)
print("Coloumns: ",df.columns)
print("Info : ",df.info())
#to find null
for i in range(25):
  print("The number of null values in column is : ",i," are : ")
  nullcheck=df.iloc[:,i].isnull().sum()
  print(nullcheck)
Y=df['default.payment.next.month']
Y
X=df.drop(['ID','default.payment.next.month'],axis='columns')
X
def normalize(X):
  X-np.mean(X)/np.std(X)
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X.iloc[:,12:24]=scaler.fit_transform(X.iloc[:,12:24])
print(X.iloc[:,12:24])
X
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.8,random_state=7)
print(X_train)
print(X_test)
print(Y_test)
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,Y_train)
Y_predicted=lr.predict(X_test)
print("First 5 predected values : ")
print(Y_predicted[0:5])
print("1st 5 tested values : ")
print(np.array(Y_test[0:5]))
print("The accuracy of logistic Regression is : ")
print(lr.score(X_test,Y_test))
from sklearn import metrics
matrix = metrics.confusion_matrix(Y_predicted, Y_test)
matrix
sns.heatmap(matrix,annot=True,cmap="Greens",fmt='0.1f')
plt.xlabel("Predicted Values :")
plt.ylabel("Test values : ")
plt.title("Confusion Matrixs for Logistic Regression")
plt.show()
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(X_train,Y_train)
Y_predicted=nb.predict(X_test)
Y_predicted
print("First 5 predected values : ")
print(Y_predicted[0:5])
print("1st 5 tested values : ")
print(np.array(Y_test[0:5]))
print("The accuracy of Accuracy of Naive_Bayes is : ")
print(nb.score(X_test,Y_test))
from sklearn import metrics
matrix = metrics.confusion_matrix(Y_predicted, Y_test)
matrix
sns.heatmap(matrix,annot=True,cmap="Reds",fmt='0.1f')
plt.xlabel("Predicted Values :")
plt.ylabel("Test values : ")
plt.title("Confusion Matrixs for Naive_bayes")
plt.show()
from sklearn.model_selection import cross_val_score
lr.fit(X,Y)
print(cross_val_score(lr,X,Y,cv=10))
np.average(cross_val_score(lr,X,Y,cv=10))
nb.fit(X,Y)
print(cross_val_score(nb,X,Y,cv=10))
np.average(cross_val_score(nb,X,Y,cv=10))
```
