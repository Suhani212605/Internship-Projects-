```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('/content/sample_data/Salary_dataset.csv')
print(df)
df.shape
print(df.describe())
X=df['YearsExperience']
print(X)
X=np.array(X)
print("In the form of array for X : ",X)
Y=df['Salary']
print(Y)
Y=np.array(Y)
print("In the form of array for Y : ",Y)
plt.scatter(X,Y)
plt.xlabel("Years Of Expereince")
plt.ylabel("Salary (in INR)")
plt.title("Introduction to Linear Regression")
plt.show()
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,)
print("X_train is : ",X_train)
print(X_train.shape)
print("Y_train is : ",Y_train)
X_train=X_train.reshape(-1,1)
print("X_train in 2D array : ",X_train)
Y_train=Y_train.reshape(-1,1)
print(X_train.shape)
print("Y_train in 2D array : ",Y_train)
X_test=X_test.reshape(-1,1)
print("X_test in 2D array : ",X_test)
Y_test=Y_test.reshape(-1,1)
print("Y_test in 2D array : ",Y_test)
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr
lr.fit(X_train,Y_train)
Y_predicted=lr.predict(X_test)
print("Y_predicted values : ",Y_predicted)
#print("Y_test values : ")
Y_test
print("Accuracy : ")
print(lr.score(X_test,Y_test))
lr.coef_
print("Intercept : ",lr.intercept_)
Y_train_predicted=lr.predict(X_train)
plt.scatter(X_train,Y_train)
plt.plot(X_train,Y_train_predicted,color="black")
plt.xlabel("Years Of Expereince")
plt.ylabel("Salary (in INR)")
plt.title("Introduction to Linear Regression")
plt.show()
```
