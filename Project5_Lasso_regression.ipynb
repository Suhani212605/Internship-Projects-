```python
#LASSO : Least Absolute Shrinkage and Selection Operator
#Reduces over fitting
'''LASSO regression is a type of regularization technique used in machine learning and statistics. It helps to:
*Reduce overfitting by adding a penalty to large coefficients, and
*Select important features by forcing some coefficients to become zero.'''
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('/content/sample_data/Salary_dataset.csv')
print(df)
print(df.shape)
print(df.describe())
X=df['YearsExperience']
print("X : ",X)
Y=df['Salary']
print("Y : ",Y)
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=7)
X=np.array(X)
print("In array for X : ",X)
plt.scatter(X,Y)
plt.xlabel('Year Of Experiemce ')
plt.ylabel('Salary')
plt.title("Introduction to lasso regression : ")
plt.show()
X_train=np.array(X_train)
print("1D array of X_train : ",X_train)
X_train=X_train.reshape(-1,1)
print("2D array of X_train : ",X_train)
Y_train=np.array(Y_train)
print("1D array of Y_train : ",Y_train)
Y_train=Y_train.reshape(-1,1)
print("Y_train in 2D array : ",Y_train)
X_test=np.array(X_test)
print("1D array of X_test : ",X_test)
X_test=X_test.reshape(-1,1)
print("X_test in 2D array : ",X_test)
Y_test=np.array(Y_test)
print("1D array of Y_test : ",Y_test)
Y_test=Y_test.reshape(-1,1)
print("Y_test in 2D array : ",Y_test)
from sklearn.linear_model import Lasso
ls=Lasso()
ls.fit(X_train,Y_train)
Y_predicted=ls.predict(X_test)
print("The predicted values are : ",Y_predicted)
print("The actucal values are : ",Y_test)
print("Accuracy : ",ls.score(X_test,Y_test))
print("Intercept : ",ls.intercept_)
Y_train_predicted=ls.predict(X_train)
plt.scatter(X_train,Y_train)
plt.plot(X_train,Y_train_predicted,color="black")
plt.xlabel("Years Of Expereince")
plt.ylabel("Salary (in INR)")
plt.title("Introduction to Lasso Regression")
plt.show()
```
