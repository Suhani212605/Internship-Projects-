```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('/content/sample_data/canada_per_capita_income.csv')
print(df)
print(df.shape)
X=df['year']
X=np.array(X)
print("The array of X : ",X)
Y=df['per capita income (US$)']
Y=np.array(Y)
print("The array of Y : ",Y)
plt.scatter(X,Y)
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.title('Canada per capita income Inrtoduction to Ridge Regression ')
plt.show()
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.85,random_state=10)
df.describe()
X_train=np.array(X_train)
print("1D array of X_train : ",X_train)
X_train=X_train.reshape(-1,1)
print("2D array of X_train : ",X_train)
Y_train=np.array(Y_train)
print("1D array of Y_train : ",Y_train)
Y_train=Y_train.reshape(-1,1)
print("Y_train in 2D array : ",Y_train)
X_test=np.array(X_test)
print("1D array of X_test : ",X_test)
X_test=X_test.reshape(-1,1)
print("X_test in 2D array : ",X_test)
Y_test=np.array(Y_test)
print("1D array of Y_test : ",Y_test)
Y_test=Y_test.reshape(-1,1)
print("Y_test in 2D array : ",Y_test)
from sklearn.linear_model import Ridge
rd=Ridge()
rd.fit(X_train,Y_train)
Y_predicted=rd.predict(X_test)
print("Y predicted value : ",Y_predicted)
print("Y test actual values : ",Y_test)
print("Score : ",rd.score(X_test,Y_test))
print("Coeficient : ",rd.coef_)
print("Intercept : ",rd.intercept_)
Y_predicted_train=rd.predict(X_train)
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.title('Canada per capita income Inrtoduction to Ridge Regression ')
plt.scatter(X_train,Y_train,color='blue')
plt.plot(X_train,Y_predicted_train,color='red')
plt.show()
arr=[2030]
arr=np.array(arr)
arr=arr.reshape(-1,1)
print(arr.shape)
print("Predicted value of 2030 : ",rd.predict(arr))
arr=[2020]
arr=np.array(arr)
arr=arr.reshape(-1,1)
arr.shape
print("Predicted value for 2020 : ",rd.predict(arr))
```
