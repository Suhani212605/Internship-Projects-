```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('/content/sample_data/canada_per_capita_income.csv')
print(df)
X=df['year']
print("X : ",X)
Y=df['per capita income (US$)']
print("Y : ",Y)
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
X_train=np.array(X_train)
print("1D array of X_train : ",X_train)
X_train=X_train.reshape(-1,1)
print("2D array of X_train : ",X_train)
Y_train=np.array(Y_train)
print("1D array of Y_train : ",Y_train)
Y_train=Y_train.reshape(-1,1)
print("Y_train in 2D array : ",Y_train)
X_test=np.array(X_test)
print("1D array of X_test : ",X_test)
X_test=X_test.reshape(-1,1)
print("X_test in 2D array : ",X_test)
Y_test=np.array(Y_test)
print("1D array of Y_test : ",Y_test)
Y_test=Y_test.reshape(-1,1)
print("Y_test in 2D array : ",Y_test)
plt.scatter(X,Y)
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.title("Introduction to Decision Tree")
plt.show()
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
dt.fit(X_train,Y_train)
Y_pred=dt.predict(X_test)
print("Y_pred test answers : ",Y_pred)
print("Y Test actual answers : ",Y_test)
print("Accuracy : ",dt.score(X_test,Y_test))
plt.scatter(X_train,Y_train,color='blue')
Y_pred_train=dt.predict(X_train)
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.title("Introduction to Decision Tree")
plt.plot(X_train,Y_pred_train,color='yellow')
print("Predicted valyue of 2030 : ",dt.predict([[2030]]))
print("Predicted valyue of 3030 : ",dt.predict([[3030]]))
```
