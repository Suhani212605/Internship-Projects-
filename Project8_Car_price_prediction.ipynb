```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('/content/sample_data/Quikr_car.csv')
print(df)
df
print(df.info())
df=df.drop(['Unnamed: 0'],axis=1)
print(df)
df['Name'].unique()
df['Name']=df['Name'].str.split(" ")
print("Name column : ",df['Name'])
df['Name']=df['Name'].str.slice(0,3).str.join(" ")
print("Name with slice : ",df['Name'])
hypenReplace=lambda x: x.replace("-",'')
df['Name']=df['Name'].apply(hypenReplace)
print("Name without hypen(-) : ",df['Name'])
print("Updated tabel : ",df)
for i in range(9):
  print("the null values in column ",i," are : ")
  nullcheck=df.iloc[:,i].isnull().sum()
  print(nullcheck)
print("unique values of location : ",df['Location'].unique())
print("Null values of location : ",df[df['Location'].isnull()])
nulllocationindex=df[df['Location'].isnull()].index
print("Null location of location column is : ",nulllocationindex)
df=df.drop(nulllocationindex)
print("Updated column is : ",df)
for i in range(9):
  print("the null values in column ",i," are : ")
  nullcheck=df.iloc[:,i].isnull().sum()
  print(nullcheck)
print("Unique values of price : ",df['Price'].unique())
df['Price']=df['Price'].str.replace(',','')
print("only price column : ",df['Price'])
df['Price']=df['Price'].str.replace('₹','')
print("only price column without ₹ : ",df['Price'])
askforprice=df[df['Price']=="Ask For Price"].index
print("The rows which have Ask For Price : ",askforprice)
df=df.drop(askforprice)
print("Updates tabel : ")
print(df)
df['Price']=pd.to_numeric(df['Price'])
print("Updated Price column : ")
print(df['Price'])
print("Minimun value of Price column : ",df['Price'].min())
print("Type:",df['Kms_driven'].dtype)
df['Kms_driven'] = df['Kms_driven'].astype(str)
print(df['Kms_driven'])
df['Kms_driven']=df['Kms_driven'].str.replace(',','')
print("After replace comma : ",df['Kms_driven'])
df['Kms_driven']=df['Kms_driven'].str.replace(' kms','')
print("After replace kms : ")
print(df['Kms_driven'])
df['Kms_driven'] = pd.to_numeric(df['Kms_driven'],errors='coerce')
print("Updated value of Kms_driven column : ",df['Kms_driven'])
df=df.drop(['Owner'],axis="columns")
print("Updated tabel is : ")
print(df)
df
for i in range(8):
  print("the null values in column ",i," are : ")
  nullcheck=df.iloc[:,i].isnull().sum()
  print(nullcheck)
labels={'PLATINUM':2,'GOLD':1}
df=df.replace(labels)
print("Updated tabel : ")
print(df)
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Location']=le.fit_transform(df['Location'])
print("Updated tabel : ")
print(df)
df['Fuel_type']=le.fit_transform(df['Fuel_type'])
print("Updated tabel : ")
print(df)
df['Name']=le.fit_transform(df['Name'])
print("Updated tabel : ")
print(df)
dummies1=pd.get_dummies(df['Location'])
print(dummies1)
df = df.drop('Location', axis=1)  # axis=1 means drop a column
print(df)
df=pd.concat([df,dummies1],axis='columns')
print("Updated tabel with location : ")
print(df)
df
df=df.drop(df.columns[45],axis=1)
print(df)
dummies2=pd.get_dummies(df['Fuel_type'])
df = df.drop('Fuel_type', axis=1)  # axis=1 means drop a column
print("Updated tabel for fuel type  drop: ")
print(df)
df=pd.concat([df,dummies2],axis='columns')
print("Updated tabel for fuel type  concat: ")
print(df)
df=df.drop(df.columns[53],axis=1)
print(df)
df=df.reset_index(drop=True)
print("Updated tabel is with correct index : ")
print(df)
sns.set_style('whitegrid')
plt.figure(figsize=(15,15))
sns.distplot(df['Price'])
plt.show()
from sklearn.preprocessing import LabelEncoder

df_encoded = df.copy()
le = LabelEncoder()
for col in df_encoded.select_dtypes(include=['object']).columns:
    df_encoded[col] = le.fit_transform(df_encoded[col])

correlation = df_encoded.corr()
print("Co relation : ")
print(correlation)
correlation
plt.figure(figsize=(15,15))
sns.heatmap(correlation,annot=True,fmt='0.2f')
plt.show()
label1=df[df['Label']==1]
print("Label 1 : ")
print(label1)
label2=df[df['Label']==2]
print("Label 2 : ")
print(label2)
sns.barplot(x=['Platinum Rated Cars','Gold Rated Cars'],y=[len(label1),len(label2)])
plt.show()
print("Grouping of comapnies : ")
df2=df.groupby(['Company']).count()
print(df2)
print("Grouping of comapnies by name descending order : ")
df3=df2.sort_values(by=['Name'],ascending=False)
print(df3)
plt.figure(figsize=(20,20))
plt.xticks(rotation=90)
sns.histplot(x=df['Company'])
plt.show()
plt.figure(figsize=(20,20))
sns.scatterplot(x=df['Kms_driven'],y=df['Price'])
plt.show()
print("table of Kms_driven below 100000 : ")
lesskms=df[df['Kms_driven']<100000]
lesskms=lesskms['Kms_driven']
print(lesskms)
print("table of Price below 1500000 : ")
lessprice=df[df['Price']<=1500000]
lessprice=lessprice['Price']
print(lessprice)
plt.figure(figsize=(20,20))
sns.scatterplot(x=lesskms,y=lessprice)
plt.show()
plt.figure(figsize=(25,20))
sns.swarmplot(x=df['Company'],y=df['Price'],size=8)
plt.show()
df3=df.sort_values(by='Price',ascending=False)
print("table of sorted price in descending : ")
print(df3)
print("table of sorted price in descending for 1st 10 value : ")
top10=df3[0:10]
print(top10)
sns.barplot(x=top10['Company'],y=top10['Price'])
plt.title("Top 10 Cars")
plt.show()
print("Distplot of year : ")
plt.figure(figsize=(20,15))
sns.distplot(df['Year'])
plt.xlabel("Year")
plt.ylabel("Density")
plt.title("Distplot for year ")
plt.show()
#feature Engineering
dummies3=pd.get_dummies(df['Name'])
print("Dummies 3 : ")
print(dummies3)
dummies4=pd.get_dummies(df['Company'])
print("Dummies 4 : ")
print(dummies4)
df=df.drop(['Name','Company'],axis="columns")
print("Updated tabel : ")
print(df)
df=pd.concat([df,dummies3,dummies4],axis='columns')
print("Updated tabel after dummies : ")
print(df)
df=df.drop(df.columns[55],axis="columns")
print("Updated tabel after droping 1: ")
print(df)
df=df.drop(df.columns[300],axis="columns")
print("Updated tabel after droping 2: ")
print(df)
def normalize(feature):
  feature=feature-np.mean(feature)/np.std(feature)
  return feature
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
kms_driven=np.array(df['Kms_driven'])
kms_driven=kms_driven.reshape(-1,1)
print("2D array of kms_driven : ")
print(kms_driven)
df['Kms_driven']=scaler.fit_transform(kms_driven)
print("Scaled array of kms_driven : ")
print(df['Kms_driven'])
print(df)
Year1=np.array(df['Year'])
Year1=Year1.reshape(-1,1)
print("2D array of Year : ")
print(Year1)
df['Year']=scaler.fit_transform(Year1)
print("Scaled array of Year : ")
print(df['Year'])
print(df)
#finding target variable Y that is dependent variable
Y=df['Price']
X=df.drop(['Price'],axis="columns")
print("X : ")
print(X)
print("Y : ")
print(Y)
#split data
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
print("X_train : ")
print(X_train)
print("X_test : ")
print(X_test)
print("Y_train : ")
print(Y_train)
print("Y_test : ")
print(Y_test)
#model selection
print("####Linear Regression#####")
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
X_train.columns = X_train.columns.astype(str)
lr.fit(X_train,Y_train)
print("coefficient : ")
print(lr.coef_)
print("intercept : ")
print(lr.intercept_)
X_test.columns = X_test.columns.astype(str)
Y_pred=lr.predict(X_test)
print("1st 5 Y_pred : ")
print(Y_pred[0:5])
print("1st 5 Y_test : ")
print(Y_test[0:5])
from sklearn.metrics import r2_score
r2=r2_score(Y_test,Y_pred)
print("R2 score : ",r2)
from sklearn.model_selection import cross_val_score
X.columns = X.columns.astype(str)  # 🔹 Fix column name type issue
cvr = cross_val_score(lr, X, Y, cv=10)
print("Cross val score : ")
print(cvr)
print("Average : ",np.average(cvr))
#lasso regression
print("####Lasso Regression#####")
from sklearn.linear_model import Lasso
ls=Lasso()
ls.fit(X_train,Y_train)
Y_predicted=ls.predict(X_test)
print("The predicted values are : ",Y_predicted)
print("The actucal values are : ",Y_test)
print("Accuracy : ",ls.score(X_test,Y_test))
print("1st 5 Y_pred : ")
print(Y_predicted[0:5])
print("1st 5 Y_test : ")
print(Y_test[0:5])
r2=r2_score(Y_test,Y_predicted)
print("R2 score for lasso: ",r2)
cvr = cross_val_score(lr, X, Y, cv=10)
print("Cross val score for lasso : ")
print(cvr)
print("Average for lasso : ",np.average(cvr))
#ridge regression
print("####Ridge Regression#####")
from sklearn.linear_model import Ridge
rd=Ridge()
rd.fit(X_train,Y_train)
Y_predicted=rd.predict(X_test)
print("Y predicted value : ",Y_predicted)
print("Y test actual values : ",Y_test)
print("Score : ",rd.score(X_test,Y_test))
print("Coeficient : ",rd.coef_)
print("Intercept : ",rd.intercept_)
print("1st 5 Y_pred : ")
print(Y_predicted[0:5])
print("1st 5 Y_test : ")
print(Y_test[0:5])
r2=r2_score(Y_test,Y_predicted)
print("R2 score for ridge: ",r2)
#Descision tree
print("####Descision Tree Regression#####")
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
dt.fit(X_train,Y_train)
Y_predicted=dt.predict(X_test)
print("Y predicted value : ",Y_predicted)
print("Y test actual values : ",Y_test)
r2=r2_score(Y_test,Y_predicted)
print("1st 5 Y_pred : ")
print(Y_predicted[0:5])
print("1st 5 Y_test : ")
print(Y_test[0:5])
print("R2 score for Decision Tree: ",r2)
from sklearn.model_selection import GridSearchCV
#for LinearRegression()

parameters = {'fit_intercept': [True, False]}
grid = GridSearchCV(lr, parameters, cv=10, scoring='r2')  # use r2 for regression
grid.fit(X, Y)
print("Best params:", grid.best_params_)
grid.cv_results_
#To find best model
import pickle
with open('CarPricePrediction.pickle', 'wb') as f:
    pickle.dump(lr, f)
#To load saved image :
with open('CarPricePrediction.pickle', 'rb') as f:
    mp = pickle.load(f)
```
